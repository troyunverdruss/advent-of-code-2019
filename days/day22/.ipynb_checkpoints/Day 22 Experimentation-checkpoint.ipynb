```python
%matplotlib inline
import matplotlib.pyplot as plt
from typing import List
import csv
```

```python
def read_raw_entries(input_file, strip=True) -> List[str]:
    lines = []
    with open(input_file, 'r', encoding='utf8') as f:
        for line in f:
            if strip:
                lines.append(line.strip())
            else:
                lines.append(line)

    return lines
```

```python
def part1_efficient(lines, cards_count, target_card, cycles, file=None):
    leading = target_card
    pos_target = target_card
    tailing = cards_count - target_card - 1

    f = open(file, 'w')
    
    
    for cycle in range(cycles):
#         print(f"Cycle {cycle}: {pos_target}")
        f.write(f"{cycle}, {pos_target}\n")
        for step in lines:
            if "deal with increment" in step:
                leading, pos_target, tailing = _deal_with_inc(cards_count, pos_target, step)
            elif "deal into new stack" in step:
                leading, tailing = tailing, leading
                pos_target = leading
            elif "cut" in step:
                count = int(step.split(" ")[1])
                # if count >= 0:
                leading = (leading - count) % cards_count
                tailing = (tailing + count) % cards_count
                pos_target = leading
            else:
                raise Exception("Unknown command: ", step)
#             print(f"Cycle {cycle + 1}, After step: {step}, target card at position: {pos_target}")
    f.close()
    return pos_target

def _deal_with_inc(cards_count, pos_target, step):
    count = int(step.split(" ")[3])
    pos_target = pos_target * count % cards_count
    leading = pos_target
    tailing = cards_count - pos_target - 1
    return leading, pos_target, tailing
```

```python
_lines = read_raw_entries("input22.txt")
part1_efficient(_lines, 10007, 2020, 20015, "data_10007_twice.csv")
part1_efficient(_lines, 107, 20, 220, "data_107_repeated_220.csv")
```

```python
xs = []
ys = []

with open ("data_10007_twice.csv", 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        xs.append(int(row[0]))
        ys.append(int(row[1]))

print(xs[:30])
print(ys[:30])

plt.plot(xs, ys, 'bo')
plt.rcParams['figure.figsize'] = [1000, 25]
plt.show()
```
